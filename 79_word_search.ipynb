```python
class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        rows = len(board)           # out of range 테스트를 위해서 필요
        cols = len(board[0])        # 아래의 함수 아래에 넣어서 매번 계산하기보다 여기에 선언하고 가는 게 나음.
        size = [[-1, 0], [1, 0], [0, -1], [0, 1]]       # size도 같은 이유로 find word에 넣지 않았음
        
        # word는 words를 리스트 형태로 사용할 수 있기에 굳이 이렇게 넣을 필요 없음.
        # word = [i for i in word]
        
        # 따로 관리하는 게 정리가 좋은 것 같음.
        def out_of_range(row, col):
            if row < 0 or row >= rows:
                return True
            if col < 0 or col >= cols:
                return True
            return False
        
        def find_word(row, col, word, i, visited):
            # position 매개변수는 알아보기 쉽도록 x, y 형태로 사용을 안했었음. 
            # 그걸 위한 거라면 row, col 형태로 넘겨주면 알아보기도 쉽고 계산을 한 번 줄일 수 있음.
            # row, col = position[0], position[1]     
            
            # 범위 확인, 원하는 목적지 확인, 방문한 장소 확인
            # visited를 사용하지 않고 board를 직접 건드리는 방식... 방문 표시에 이점은 있으나, 코드 오류나면 오히려 힘들어지던데..
            # 별로 좋아하지 않음. 원본은 되도록 유지하는 것이 개인 방침인데, 교수님한테 물어볼 것.
            # i >= len(word) 빼도 다른 조건에서 거르기 가능. 아닌데.. word[i]에서 인덱스 에러 위험 존재. 
            # 완료 조건을 위에 넣으면 위의 조건 삭제 가능?
            # 솔루션을 보면 길이가 맞아서 트루로 돌려보내도 윗 단계에서 범위와 알파벳 일치 검사를 다시 하기 때문에 넘어감.
            if out_of_range(row, col) or i >= len(word) or board[row][col] != word[i] or [row, col] in visited:
                return False
            elif i == len(word) -1 and board[row][col] == word[-1]:
                return True
            else:
                # 디버깅 관점에서도 visited를 따로 관리하는 게 효율적이라 생각됨.
                visited.append([row, col])
            
            # 완료 조건
            if board[row][col] == word[-1] and i == len(word) - 1:
                return True
            
            for x_move, y_move in size:
                x, y = row + x_move, col + y_move
                if find_word([x, y], word, i+1, visited):
                    return True
            visited.pop()
            return False
                
        positions = []
        for row in range(rows):
            for col in range(cols):
                if board[row][col] == word[0]:
                    positions.append([row, col])
        
        
        i = 0
        for position in positions:
            visited = []
            check= find_word(position, word, i, visited)
            if check:
                return True
        
        return False
```
